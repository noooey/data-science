```python
import torch
import matplotlib.pyplot as plt

Z = torch.randn(200, 2)

cov = torch.tensor([[1, 0.2], [0.2, 0.5]])

Z = Z @ cov

plt.scatter(Z[:,0], Z[:,1])
plt.axis("equal")
plt.show()
```

```python
def findPC(Z):
  dim = Z.shape[1]   # z.shape = 200, 2
  w = torch.randn(dim)
  w = w / ((w * w).sum() ** 0.5)
  
  lr = 0.1

  for epoch in range(101):
    w.requires_grad_(True)
    variance = ((Z @ w.view(-1, 1)) ** 2).mean()

    variance.backward()

    with torch.no_grad():
      w = w + lr * w.grad
      w = w / ((w **2).sum() ** 0.5)
    
  return w, variance.item()
```

```python
def PCA(Z, n):
  W = []
  V = []
  
  for _ in range(n):
    w, v = findPC(Z)
    W.append(w)
    V.append(v)
    Z = Z - (Z @ w.view(-1, 1)) * w
  
  return W, V
```

```python
def transform(Z, W):
  return Z @ torch.stack(W).T
```

```python
W, V = PCA(Z, 2)

plt.axis("equal")
plt.scatter(Z[:,0], Z[:,1])
plt.plot([0, W[0][0]], [0, W[0][1]], c="red") # PC1
plt.plot([0, W[1][0]], [0, W[1][1]], c="green") # PC2
plt.show()

Zt = transform(Z, W)
plt.scatter(Zt[:,0], Zt[:,1])
plt.axis("equal")
plt.show()
```

```python
import requests

r = requests.get("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data")
with open("iris.data", "wb") as f:
  f.write(r.content)
```

```python
vectors = []
answers = []

for line in open("iris.data", "r"):
  try:
    items = line.strip().split(",")
    vectors.append([float(i) for i in items[:4]])
    answers.append(items[4])
  except:
    pass

print(vectors)
print(answers)
species = {a: i for i, a in enumerate(set(answers))}
species
```

```python
X = torch.tensor(vectors)
Z = X - X.mean(dim=0)
```

```python
W, V = PCA(Z, 4)
V
```

```python
# scree plot

plt.plot(range(1, 5), V, "-o")
plt.show()
```

```python
Zt = transform(Z, W)

plt.scatter(Zt[:,0], Zt[:,1], c=[species[a] for a in answers], alpha=0.3)
plt.axis("equal")
plt.show()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
Zt2 = pca.fit_transform(X)

plt.scatter(Zt2[:,0], Zt2[:,1], c=[species[a] for a in answers], alpha=0.3)
plt.axis("equal")
plt.show()
```
